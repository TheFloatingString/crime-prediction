```python
import pandas as pd 
import numpy as np
import datetime
```

```python
nasdaq_df = pd.read_csv("../datasets/NASDAQ_DATA.csv", index_col="Date").drop(columns=['Adj Close'])
print(nasdaq_df)
```

```python
base = datetime.datetime.strptime("2015-01-01", "%Y-%m-%d")
end = datetime.datetime.strptime("2019-07-06", "%Y-%m-%d")
delta = end - base
unique_PI_dates = [str(base + datetime.timedelta(days=x)).split(' ')[0] for x in range(delta.days + 1)]
print(len(unique_PI_dates))
```

```python
nasdaq_df = nasdaq_df.dropna()
for column in nasdaq_df.columns:
    nasdaq_df[column + '_variance'] = nasdaq_df[column].pct_change()
    
nasdaq_df = nasdaq_df.fillna(0)

print(nasdaq_df.shape)
```

```python
def nearest(items, pivot):
    return min(items, key=lambda x: abs(x - pivot))
```

```python
available_dates = nasdaq_df.index.to_list()
datetime_dates = [datetime.datetime.strptime(x, "%Y-%m-%d") for x in available_dates]
print(len(available_dates))

for date_code in unique_PI_dates:
    if date_code in available_dates: continue
    elif date_code not in available_dates:
        closest_date = str(nearest(datetime_dates, datetime.datetime.strptime(date_code, "%Y-%m-%d"))).split(' ')[0]
        nasdaq_df.loc[date_code] = nasdaq_df.ix[closest_date]

nasdaq_df = nasdaq_df.sort_index()
X_train_nasdaq = nasdaq_df.to_numpy()

print(X_train_nasdaq.shape)
np.save('X_train_nasdaq.npy', X_train_nasdaq)
```
