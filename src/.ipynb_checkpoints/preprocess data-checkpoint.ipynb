```python
import pandas as pd
import json
import csv
import numpy as np
from datetime import datetime, timedelta
```

```python
f = open("interventionscitoyendo.csv", 'r')
```

```python
df_crimes = pd.read_csv("interventionscitoyendo.csv",delimiter=',', encoding = "ISO-8859-1").dropna()
```

```python
df_crimes.shape
```

```python
df_crimes.head()
```

```python
df_crimes = df_crimes.sort_values(by=["DATE"])
```

```python
df_crimes.head()
```

```python
base = datetime(2015,1,1)
x = 1
numdays=1648
date_list = [base + timedelta(days=x) for x in range(0, numdays)]
```

```python
date_list[-1]
```

```python
df_pdq = df_crimes.loc[df_crimes["PDQ"]==33.0]
print(df_pdq.shape)
df_pdq.head(50)
```

```python
types_of_crimes = list(set(df_crimes["CATEGORIE"].values))
phase_of_day = list(set(df_crimes["QUART"].values))
print(types_of_crimes)
```

```python
new_data_dict = {}
for datetime_obj in date_list[:10]:
    selected_rows = df_pdq.loc[df_pdq["DATE"]==datetime_obj.strftime("%Y-%m-%d")]
    initial_crime_vector = [0.0]*6
    initial_phase_of_day_vector = [0.0]*3
    for index, row in selected_rows.iterrows():
        initial_crime_vector[types_of_crimes.index(row["CATEGORIE"])] += 1.0
        initial_phase_of_day_vector[phase_of_day.index(row["QUART"])] += 1.0
    new_data_dict[datetime_obj.strftime("%Y-%m-%d")] = initial_crime_vector + initial_phase_of_day_vector
```

```python
new_data_dict
```

```python
# actual loop
```

```python
pdq_numbers = list(set(df_crimes["PDQ"]))
print(pdq_numbers)

# for each police station
for pdq_number in pdq_numbers:
    df_pdq = df_crimes.loc[df_crimes["PDQ"]==pdq_number]
    
    indv_pdq_data_dict = {}
    
    crime_category_dict = {}
    for category in types_of_crimes+phase_of_day:
        crime_category_dict[category] = []
    
    # for each day
    for datetime_obj in date_list:
        
        selected_rows = df_pdq.loc[df_pdq["DATE"]==datetime_obj.strftime("%Y-%m-%d")]
        initial_crime_vector = [0.0]*6
        initial_phase_of_day_vector = [0.0]*3
        for index, row in selected_rows.iterrows():
            initial_crime_vector[types_of_crimes.index(row["CATEGORIE"])] += 1.0
            initial_phase_of_day_vector[phase_of_day.index(row["QUART"])] += 1.0
        indv_pdq_data_dict[datetime_obj.strftime("%Y-%m-%d")] = initial_crime_vector + initial_phase_of_day_vector
#         list_of_pdq_dicts.append(indv_pdq_data_dict)

        for category, item in zip(list(crime_category_dict.keys()), initial_crime_vector+initial_phase_of_day_vector):
            crime_category_dict[category].append(item)
        
    df_ind_pdq = pd.DataFrame(data=crime_category_dict, index=date_list)
    df_ind_pdq.to_csv("pdq_"+str(int(pdq_number))+".csv", encoding='utf-8')
    print(pdq_number)
```
